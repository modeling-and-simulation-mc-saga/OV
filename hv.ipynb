```python
import matplotlib.pyplot as plt
import math
from typing import Callable
import sys
sys.path.append("../lib")
from pylib import pltInit,Point,DataFile
```

```python
def drawHV(fType:str,xMax:float,yMax:float):
    plt.figure(figsize=(9,6),facecolor='white')
    
    plt.xlim(0,xMax)
    plt.ylim(0,yMax)
    plt.xlabel('$\\Delta x$')
    plt.ylabel('$v$')
    xl, yl = DataFile(fType+'-hv.txt').readData()
    plt.scatter(xl,yl)
    plt.savefig(fType+'-hv.pdf')
```

```python
pltInit()
drawHV('step',20,15)
drawHV('Tanh',40,40)
```
